```
# Manipulação de dados (com pandas)

# Concatenando dados (criando novos conjuntos, combinando dados de diferentes fontes)

import pandas as pd
import numpy as np

from pandas import DataFrame
```

```
df = pd.DataFrame(np.arange(36).reshape(6,6))
df
```

```
df2 = pd.DataFrame(np.arange(15).reshape(5,3))
df2
```

```
#para concatenar dataframes usa-se o método concat()

pd.concat([df,df2])

# note abaixo que basicamente pegou as linhas do segundo objeto e colocou embaixo das linhas do primeiro objeto.
# para pegar por coluna usar o axis =1
# veja no exemplo a seguir: 
```

```
pd.concat([df,df2], axis=1)

```
